```python
import pandas as pd 
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
df = pd.read_csv("./steel_until_2021.csv", sep='\t')
#df.set_index('Date', inplace=True)
#df.sort_index(ascending=False)
df.columns
```

```python
df.Date = pd.to_datetime(df.Date).dt.date
```

```python
import datetime 
df = df[(df['Date']>datetime.date(2007,1,1)) & (df['Date']<datetime.date(2022,1,15))]  
```

```python
df.groupby('name')['value'].plot(legend=True, figsize=(28, 26) )
```

```python
  
dfp = df.pivot_table(index='Date', columns='name', values='value')#, aggfunc='sum'

# plot
ax = dfp.plot(figsize=(20, 15))
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```
